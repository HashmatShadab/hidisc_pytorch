```python
import json
import matplotlib.pyplot as plt

filename = "./log.txt"

# List of keys to be plotted
keys = ['train_total_loss', 'train_clean_loss', 'train_adv_loss', "train_avg_increase"]

epochs = []

# Create a dictionary to store values for each key
values = {key: [] for key in keys}

with open(filename, 'r') as file:
    for line in file:
        data = json.loads(line)

        epochs.append(data['Epoch'])

        # Append each key's value to its corresponding list
        for key in keys:
            values[key].append(data[key])

# Create subplots
fig, axs = plt.subplots(1, len(keys), figsize=(20, 10))
fig.suptitle('Plot of Keys vs Epoch')

# For each key, plot its values
for i, key in enumerate(keys):
    axs[i].plot(epochs, values[key])
    axs[i].set(xlabel='Epoch', ylabel=key)

plt.tight_layout()
plt.show()
```
